## **Filtering our data**

Pandas has various methods and techniques to filter and query our data.


### **df.query()**

The `query()` method is very powerful and can be extended and combined to filter the dataframe in any number of ways.

Here we are looking for rows that have 1543 as a Year value:

> Note the == double equal operator.

> **See more on operators at the end of this notebook.**

```python
df.query('Year == 1543')
```

The `<` operator will return any rows where `Year` is "smaller" than 1800.

```python
df.query('Year < 1800')
```

### **Chaining Pandas methods**

Pandas methods can be used one after the other in a single execution.

Here we are querying the dataframe for rows where `Year` is before 1800 and we are just looking at the `Entity` column and seeing what unique values there are in that subset.

```python
df.query('Year < 1800')['Entity'].unique()
```

### **df['series].nunique()**

This simply returns the count of the uniques for a series.

```python
df.query('Year < 1800')['Entity'].nunique()
```

### **More ways to use query()**

When searching for strings, we use quotes to isolate the string.

```python
df.query('Entity == "Namibia"')
```

Query with **multiple** criteria:

Here we are searching for rows where `Entity` is "Namibia" and `Year` is 2010.  

```python
df.query('Entity == "Namibia" & Year == 2010')
```

Queries can be as simple or complicated as you like:

Here we are searching for rows where `Year` is 2010 and `Entity` is "Namibia" OR (|) rows where `Year` is 2010 and `Entity` is "Zimbabwe".

```python
df.query('Year == 2010 & Entity == "Namibia" | Year == 2010 & Entity == "Zimbabwe"')
```

This could also be rewritten as:

> Note the brackets around our `Entity` query. The brackets ensure that that is considered as a single concept.

```python
df.query('(Entity == "Namibia" | Entity == "Zimbabwe") & Year == 2010')
```

So we can also read multiple years for instance:

```python
df.query('(Entity == "Namibia" | Entity == "Zimbabwe") & Year > 1950')
```

### **More ways to filter data**

The `query()` method is very useful, but there are other ways to query the data. We can use series queries directly:

```python
df[df['Year'] == 1543]
```

Or with multiple criteria.

> Note the brackets used to separate criteria.

```python
df[(df['Entity'] == 'Namibia') & (df['Year'] == 2010)]
```

> Again, note the use of the brackets to separate criteria.

```python
df[((df['Entity'] == 'Namibia') | (df['Entity'] == 'Zimbabwe')) & (df['Year'] == 2010)]
```

## **Plotting the data**

Plotting our datasets is essential to understand and analyse our datasets.

Pandas `plot()` defaults to line, but can produce a variety of chart types:

### **Fixing bad charts**

Before plotting data - especially timeseries data - it is essential to sort the data. Consider this problem below:

Here we are querying the data for `Entity` == "United Kingdom", but the chart looks very strange.

```python
df.query('Entity == "United Kingdom"').plot(x='Year', y='LifeExpectancy')
```

If we investigate the queried data, we can see the years are not in order:

```python
df.query('Entity == "United Kingdom"')
```

### **df.sort_values('series')**

Here we sorting the values before plotting.

> Remember at no point have we actually made any changes to our dataframe. These queries and sortings are done at runtime (when executing the cell) and doesn't affect the original dataframe.

> At any point in time, we **CAN** set our query to a new variable or update our original variable. ie `df = df.qeury('Entity == "Namibia")` will mean the `df` will only now contain "Namibia".

> This might be very useful to store different versions of the data and save time (and complexity) on retyping queries.

```python
df.query('Entity == "United Kingdom"').sort_values('Year').plot(x='Year', y='LifeExpectancy')
```

### **Combining plots**

There is more than one way to combine plots. The first method is essentially this:

- Create a plot and store it in a variable.
- Make a second plot and use the first plot as the axes. (`ax=base_chart`)

```python
base_chart = df.query('Entity == "United Kingdom"').sort_values('Year').plot(x='Year', y='LifeExpectancy')
df.query('Entity == "Namibia"').sort_values('Year').plot(x='Year', y='LifeExpectancy', ax=base_chart)
```

We can change our queries to line up the charts a bit better:

```python
base_chart = df.query('Entity == "United Kingdom" & Year > 1949').sort_values('Year').plot(x='Year', y='LifeExpectancy', label="UK")
df.query('Entity == "Namibia" & Year > 1949').sort_values('Year').plot(x='Year', y='LifeExpectancy', ax=base_chart, label="Namibia")
```

The problem with this approach is that we would need to create a chart for every Entity and this might not be possible when comparing more variables.

### **df.pivot()**

A better way to approach charting on a dataset might be to pivot our data.

> Just like in a spreadsheet, we choose our index, columns and values.


```python
df_pivot = df.pivot(index='Year', columns='Entity', values='LifeExpectancy')
df_pivot
```

```python
df_pivot.plot(legend=False)
```

Let's filter it to our selected countries and time range:

```python
df_pivot = df.query('(Entity == "United Kingdom" | Entity == "Namibia") & Year > 1949').pivot(index='Year', columns='Entity', values='LifeExpectancy')
df_pivot
```

```python
df_pivot.plot()
```

### **Using variables in filters**

Suppose we wanted to have more countries, we could make the filtering a bit cleaner and simpler like this:

```python
countries = ["United Kingdom", "Namibia", "South Africa", "Nigeria", "Kenya"]
df_pivot = df.query('Entity == @countries & Year > 1949').pivot(index='Year', columns='Entity', values='LifeExpectancy')
df_pivot
```

```python
df_pivot.plot()
```

### **Styling the plot**

```python
df_pivot_plot = df_pivot.plot(title="Life Expectancy", ylabel="Life Expectancy (Years)", fontsize=8, figsize=(12, 6))
df_pivot_plot.grid(color='#eee', linestyle='--', linewidth=0.5)
df_pivot_plot.set_facecolor('#fefefe')
```

### **Adding reference lines**

Use the `axvline()` or `axhline()` method to add a reference line.

```python
df_pivot_plot = df_pivot.plot(title="Life Expectancy", ylabel="Life Expectancy (Years)", fontsize=8, figsize=(12, 6))
df_pivot_plot.grid(color='#eee', linestyle='--', linewidth=0.5)
df_pivot_plot.set_facecolor('#fefefe')
df_pivot_plot.axvline(x=2002, color='#333', linestyle='--', linewidth=0.8)
df_pivot_plot.axhline(y=55, color='#333', linestyle='--', linewidth=0.8)
```

### **More styling**

You can change the colour of individual lines by adding a `color` parameter to the `plot()`

```python
df_pivot_plot = df_pivot.plot(title="Life Expectancy", ylabel="Life Expectancy (Years)", fontsize=8, figsize=(12, 6), color=['grey','grey','grey','red'])
df_pivot_plot.grid(color='#eee', linestyle='--', linewidth=0.5)
df_pivot_plot.set_facecolor('#fefefe')
```

### **Conditional styling with code**

We might need to style individual lines and can potentially do it like this:

```python
colors = ['red' if country == "Nigeria" else 'grey' for country in df_pivot.columns]
df_pivot_plot = df_pivot.plot(title="Life Expectancy", ylabel="Life Expectancy (Years)", fontsize=8, figsize=(12, 6), color=colors)
```

## **Combining datasets**

The ability to combine datasets is essential for any meaningful data analysis or exploration. This is also one of Pandas key strengths and it is perhaps even better at it than your usual spreadsheet.

### **Load another dataset to combine**

Let's load another set and examine it.

```python
gdp_df = pd.read_csv('national_gdp.csv')
```

```python
gdp_df
```

```python
gdp_df.describe()
```

### **df.merge()**

The Pandas `merge()` method allows us to merge datframes on one or more columns.

Let's look at our original set again:


```python
df
```

### **Merging methods**

We want to merge on Code and Year, so we can specify both columns in the `on` parameter. We are choosing `left` for the `how` parameter. This means it will keep everything in the left dataframe (in this case `df` and add to it values from the right dataframe (`gdp_df`) where possible.

![Joins](https://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png)


```python
merged_df = pd.merge(df, gdp_df, on=['Code', 'Year'], how='left')
merged_df
```

### **Fixing a bad merge**

The merged dataframe has **34795 rows**. When doing a left merge we would expect it to have exactly the same amount of rows as the left dataframe - **20755**. So what went wrong?

Here we can see that there are 20755 rows, but not all of them are **non-null** for the `Code` column.

Similarly on `gdp_df` we can see there are quite a few **non-null** entries for the `Code` column.

the problem with that is that we can't merge on **nulls** as they will match multiple times. Ending up with a bigger dataframe as we saw.

```python
df['Code'].info()
```

```python
gdp_df['Code'].info()
```

### **df['series].isnull()**

In this case, let's see what Entities have Codes with **null**:

```python
df[df['Code'].isnull()]
```

```python
df[df['Code'].isnull()]['Entity'].unique()
```

```python
gdp_df[gdp_df['Code'].isnull()]
```

```python
gdp_df[gdp_df['Code'].isnull()]['Entity'].unique()
```

### **Dropping the null rows**

Let's get rid of all the null rows with a `notnull()` method.

Here we are making a new datafame and saying it is everthing where Code is not null.

> We're making a new variable so that we can always get back to the original `df` if we want to.

```python
df_filtered = df[df['Code'].notnull()]
df_filtered
```

We'll do the same for the other dataframe:

```python
gdp_df_filtered = gdp_df[gdp_df['Code'].notnull()]
gdp_df_filtered
```

### **df.merge()**

Now the merge should work:

```python
merged_df = pd.merge(df_filtered, gdp_df_filtered, on=['Code', 'Year'], how='left')
merged_df
```

### **Renaming columns**

Let's rename the `GDP (constant 2015 US$)` column.

```python
merged_df.rename(columns={'GDP (constant 2015 US$)': 'GDP'}, inplace=True)
merged_df
```

### **Dropping columns**

Let's get rid of the second Entity_y column:

```python
merged_df.drop(columns=['Entity_y'], inplace=True)
merged_df.rename(columns={'Entity_x': 'Entity'}, inplace=True)
merged_df
```

### **Plotting the data**

Let's see every country in the dataset's life-expectancy against their GDP in 2020

```python
merged_df.query('Year == 2020').plot(kind='scatter',
    x='GDP',
    y='LifeExpectancy',
    alpha=0.7,
    figsize=(10, 6),
    logx=True,
    title='Life Expectancy vs. GDP in 2020',
    xlabel='GDP (constant 2015 US$)',
    ylabel='Life Expectancy')
```

### **Plotting the data with Plotly**

Pandas uses Matplotlib as its default plotting library, but there are many other plotting libraries, each with different strengths.

> Plotting libraries include: [Matplotlib](https://matplotlib.org/), [Seaborn](https://seaborn.pydata.org/), [Plotly](https://dash.plotly.com/), [Plotnine](https://plotnine.org/), [Bokeh](https://bokeh.org/) and [Altair](https://altair-viz.github.io/).

```python
import plotly.express as px

fig = px.scatter(merged_df.query('Year == 2020'),
          x='GDP',
          y='LifeExpectancy',
          hover_name='Entity',
          log_x=True,
          title='Life Expectancy vs. GDP in 2020',
          labels={
              'GDP': 'GDP (constant 2015 US$)',
              'LifeExpectancy': 'Life Expectancy'
          },
          width=1000,
          height=800)
fig.show()
```

### **Plotting the data with Bokeh**

Here's the same chart using Bokeh.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

output_notebook()

p = figure(title='Life Expectancy vs. GDP in 2020',
           x_axis_label='GDP (constant 2015 US$)',
           y_axis_label='Life Expectancy',
           x_axis_type="log",
           width=1000,
           height=800)

scatter = p.scatter(x='GDP', y='LifeExpectancy', source=merged_df.query('Year == 2020'),
                    size=10, alpha=0.7)

hover = HoverTool(tooltips=[("Entity", "@Entity")], renderers=[scatter])
p.add_tools(hover)

show(p)
```
